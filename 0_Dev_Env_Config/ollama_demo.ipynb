# Ollama Demo Notebook

Uses llama3.2:3b running locally on Ollama.

**Prerequisites:**
- Ensure [Ollama installed](https://ollama.ai/download) is running by running: `ollama serve` in your Terminal
- Pull the llama3.2:3b model by running: `ollama pull llama3.2:3b` in your Terminal

```python
# Step 1: Install the ollama Python library
import ollama
```

```python
# Step 2: Make equest to llama3.2:3b
# No API key needed - it runs entirely on your local machine

response = ollama.chat(
    model='llama3.2:3b',
    messages=[
        {
            'role': 'user',
            'content': 'Who are you?'
        }
    ]
)

# Step 3: Display the response
print(response['message']['content'])
```
